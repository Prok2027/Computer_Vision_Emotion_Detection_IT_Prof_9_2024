```python
import torch
import pandas as pd
import os
from torchvision.transforms.v2 import Compose, PILToTensor, ToDtype, Normalize, RandomHorizontalFlip
from torch.utils.data import DataLoader, ConcatDataset, random_split
from PIL import Image
from torchvision import transforms
from concurrent.futures import ThreadPoolExecutor
```

```python
df1 = pd.read_excel("drive/MyDrive/dataset_1_no_augmentation.xlsx")
df2 = pd.read_excel("drive/MyDrive/dataset_2_no_augmentation.xlsx")
```

```python
transform_base = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
])
```

```python
ts = (64, 64)
```

```python
images = []
with ThreadPoolExecutor() as executor:
    fut = [executor.submit(lambda p: transform_base(Image.open(p).convert('RGB').resize(ts)), img_path) for img_path in df2['file_path']]
    for fut in futs:
        images.append(fut.result())
```

```python
df2['img'] = images
```

```python
df2.head()
```

```python
total = len(df1['file_path'])
```

```python
images = []
with ThreadPoolExecutor() as executor:
    futs = []
    for i, img_path in enumerate(df1['file_path']):
        futs.append(executor.submit(lambda p: transform_base(Image.open(p).convert('RGB').resize(ts)), img_path))
        print(f"Загружается изображение {i + 1} из {total}...")

    for fut in futs:
        images.append(fut.result())
```

```python
df1['img'] = images
```
