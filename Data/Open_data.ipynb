```python
import torch
import os
import pandas as pd
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
emotions = ['Anger.jpg', 'Contempt.jpg', 'Disgust.jpg',
            'Fear.jpg', 'Happy.jpg', 'Neutral.jpg',
            'Sad.jpg', 'Surprised.jpg']
```

```python
data = []
for folder in range(19):
    fpath = os.path.join('drive/MyDrive/archive/images', str(folder))
    if os.path.exists(fpath):
        for emotion in emotions:
            filep = os.path.join(fpath, emotion)
            if os.path.exists(filep):
                data.append((filep, emotion))
df = pd.DataFrame(data, columns=['file_path', 'emotion'])
print(len(df))
df.head()
```

```python
e_k = {'Anger.jpg': 0, 'Contempt.jpg': 1, 'Disgust.jpg': 2,
            'Fear.jpg': 3, 'Happy.jpg': 4, 'Neutral.jpg': 5,
            'Sad.jpg': 6, 'Surprised.jpg': 7}
```

```python
for i in range(len(df)):
  df.loc[i, 'emotion'] = e_k.get(df['emotion'][i], -1)
```
